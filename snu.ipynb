```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url_SNU_sc = 'https://data.statistics.sk/api/v2/dataset/nu0004qs/all/1.Q.,2.Q.,3.Q.,4.Q./UKAZ06,UKAZ07,UKAZ08,UKAZ09,UKAZ13,UKAZ14,UKAZ15/METO02/MJ01?lang=sk&type=csv'
```

```python
SNU_sc = pd.read_csv(url_SNU_sc, sep=';', skiprows=7, header=0, names=['Rok','Kvartal','Kod','Kod2', 'Kod3', 'Hodnota'])
```

```python
SNU_sc['obdobie'] = SNU_sc['Rok'].astype(str)+' '+SNU_sc['Kvartal']
```

```python
SNU_sc
```

```python
df = SNU_sc.pivot(index='obdobie', columns='Kod', values='Hodnota').rename(columns={
    "UKAZ06": "spotreba domácností",
    'UKAZ07': 'spotreba neziskových inštitúcií slúžiacim domácnostiam',
    'UKAZ08': 'spotreba vlády',
    'UKAZ09': 'hrubá tvorba kapitálu',
    'UKAZ13': 'vývoz',
    'UKAZ14': 'dovoz',
    'UKAZ15': 'HDP',
})
```

```python
df
```

```python
df.plot(figsize=(20,10))
```
